```python
%matplotlib inline 

import json
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
folder_name = "results/CartPole-v0/"
file_name = "tb_train_episode_reward"
with open(folder_name + file_name + ".json", 'r') as fh:
    data1 = json.load(fh)
    data1 = np.array(data1)
file_name = "tb_train_eval_reward"
with open(folder_name + file_name + ".json", 'r') as fh:
    data2 = json.load(fh)
    data2 = np.array(data2)

# plt.title('CartPole: Episode Reward during Training')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.scatter(data1[:,1], data1[:,2], label='Training', s=5, color='black')
plt.scatter(data2[::10,1], data2[::10,2], label='Evaluation', s=15, color='orange')
# plt.plot(data1[:,1], data1[:,2], label='Training')
# plt.plot(data2[:,1], data2[:,2], label='Evaluation', color='orange')
plt.legend()
plt.grid()
plt.savefig(folder_name + "tb_train_eval_reward.pdf")
```

```python
folder_name = "results/CartPole-v0/"
file_name = "tb_test_episode_reward_solved"
with open(folder_name + file_name + ".json", 'r') as fh:
    data1 = json.load(fh)
    data1 = np.array(data1)

file_name = "tb_test_episode_reward_fullrun"
with open(folder_name + file_name + ".json", 'r') as fh:
    data2 = json.load(fh)
    data2 = np.array(data2)

# plt.title('CartPole: Episode Reward for Testing')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.scatter(data1[:,1], data1[:,2], label='Solved', marker='o', s=50)#, markersize=11)
# plt.plot(data2[:,1], data2[:,2], label='Full Run', marker='X', linewidth=0, markersize=8)
plt.plot(range(15), np.mean(data1[:,2])*np.ones(15), label="mean Solved")
plt.scatter(data2[:,1], data2[:,2], label='Full Run', marker='X', s=20)#, s=8)
plt.plot(range(15), np.mean(data2[:,2])*np.ones(15), label="mean Full Run")
plt.grid()
plt.legend(bbox_to_anchor=(1, 0.8))
plt.savefig(folder_name + "tb_test_reward" + ".pdf")
```

```python
folder_name = "results/CarRacing-v0/HistOf0/"
file_name_a = "tb_train_HistOf0_"
file_name_b = "_episode_reward"

file_names_1 = []
for i in range(7):
    file_names_1.append(file_name_a + str(i+1) + file_name_b)
# print(file_names_1)

file_name_b = "_eval_reward"

file_names_2 = []
for i in range(7):
    file_names_2.append(file_name_a + str(i+1) + file_name_b)
# print(file_names_2)
    

episode_rewards = []

for file_name in file_names_1:
    with open(folder_name + file_name + ".json", 'r') as fh:
        episode_reward = json.load(fh)
        episode_reward = np.array(episode_reward)
    episode_rewards.append(episode_reward)
    
    
eval_rewards = []

for file_name in file_names_2:
    with open(folder_name + file_name + ".json", 'r') as fh:
        eval_reward = json.load(fh)
        eval_reward = np.array(eval_reward)
    eval_rewards.append(eval_reward)
    

episode_reward_values = np.array([])
for episode_reward in episode_rewards:
    episode_reward_values = np.append(episode_reward_values, episode_reward[:,2])

eval_reward_values = np.array([])
for eval_reward in eval_rewards:
    eval_reward_values = np.append(eval_reward_values, eval_reward[:,2])
    

plt.xlabel('Episode')
plt.ylabel('Reward')

print(len(episode_reward_values))


    # plt.scatter(range(len(episode_reward_values)), episode_reward_values[:], s=8)
    # plt.plot(range(len(eval_reward_values)), eval_reward_values[:])#, s=5)
plt.scatter(range(len(episode_reward_values)), episode_reward_values[:], label='Training', s=5, color='black')
plt.scatter(range(len(eval_reward_values))[::10], eval_reward_values[::10], label='Evaluation', s=15, color='orange')
# plt.plot(data1[:,1], data1[:,2], label='Training')
# plt.plot(data2[:,1], data2[:,2], label='Evaluation', color='orange')
plt.legend()
plt.grid()
plt.savefig(folder_name + "tb_train_eval_reward.pdf")
```
