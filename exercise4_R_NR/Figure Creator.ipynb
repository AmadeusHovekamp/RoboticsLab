```python
%matplotlib inline 

import json
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
folder_name = "results/CartPole-v0/"
file_name = "tb_train_episode_reward"
with open(folder_name + file_name + ".json", 'r') as fh:
    data1 = json.load(fh)
    data1 = np.array(data1)
file_name = "tb_train_eval_reward"
with open(folder_name + file_name + ".json", 'r') as fh:
    data2 = json.load(fh)
    data2 = np.array(data2)

plt.title('CartPole: Episode Reward during Training')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.plot(data1[:,1], data1[:,2], label='Training')
plt.plot(data2[:,1], data2[:,2], label='Evaluation')
plt.grid()
plt.savefig(folder_name + file_name + ".pdf")
```

```python
folder_name = "results/CartPole-v0/"
file_name = "tb_test_episode_reward_solved"
with open(folder_name + file_name + ".json", 'r') as fh:
    data1 = json.load(fh)
    data1 = np.array(data1)

file_name = "tb_test_episode_reward_fullrun"
with open(folder_name + file_name + ".json", 'r') as fh:
    data2 = json.load(fh)
    data2 = np.array(data2)

plt.title('CartPole: Episode Reward for Testing')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.plot(data1[:,1], data1[:,2], label='Solved')
plt.plot(data2[:,1], data2[:,2], label='Full Run')
plt.grid()
plt.legend()
plt.savefig(folder_name + "tb_test_episode_reward" + ".pdf")
```
